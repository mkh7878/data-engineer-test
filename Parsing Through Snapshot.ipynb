```python
import re
import pandas as pd

# Initialize an empty list to store the parsed records
parsed_records = []

# Read the data from the 'officer_snapshot.txt' file
with open('officer_snapshot.txt', 'r') as file:
    lines = file.readlines()

# Define a regular expression pattern to extract the first 3 items from each line
pattern = r'(\d{8})(\d{12})\s+(\d{8})'

# Loop through the lines and extract the first 3 items using the regular expression
for line in lines:
    match = re.search(pattern, line)
    if match:
        data = match.groups()
        # print(data)  # Add this line to inspect what's being captured
        # Create a dictionary with field names
        record = {
            "Company Number": data[0],
            "Person Number": data[1],
            "Date Appointment": data[2],
        }
        parsed_records.append(record)

# Check if any records were parsed
if parsed_records:
     records_df = pd.DataFrame(parsed_records)
    # Display the resulting DataFrame
    # print(df)
else:
    print("No records were parsed.")
```

```python
# Initialize an empty list to store the parsed records
strings_list = []

# Read the data from the 'officer_snapshot.txt' file
with open('officer_snapshot.txt', 'r') as file:
    lines = file.readlines()

# Define a regular expression pattern to extract an 8-digit number, text, and <
pattern3 = r'(\d{8})\s+([\w\s.]+)<'

# Loop through the lines and extract the desired information using the third pattern
for line in lines:
    match = re.search(pattern3, line)
    if match:
        data = match.groups()
        # Create a dictionary with field names for this pattern
        record = {
            "String": data[1],
        }
        strings_list.append(record)  # Append the record to the list
        
    else:
        print(line)

# Check if any records were parsed
if strings_list:
    stdf = pd.DataFrame(strings_list)
    # Display the resulting DataFrame
    print(stdf)
else:
    print("No records were parsed.")
```

```python

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(strings_list)

def extract_longest_consecutive_numbers(input_string):
    import re
    # Remove the 8-digit number at the beginning of the string
    input_string = re.sub(r'^\d{8}', '', input_string)
    
    # Remove leading whitespace while preserving whitespace after the first letter
    input_string = re.sub(r'^\s+', '', input_string)
    
    # Find all consecutive sequences of digits in the modified string
    digit_sequences = re.findall(r'\d+', input_string)
    if not digit_sequences:
        return (None, None, None)
    longest_sequence = max(digit_sequences, key=len)
    start_index = input_string.index(longest_sequence)
    end_index = start_index + len(longest_sequence)
    
#     if len(longest_sequence) < 6:
#         print(input_string)
#         dob = ''
        
#     else:
#         dob = longest_sequence[-6:]

    dob = ''
    title = None
    postcode = ''
    
    if len(longest_sequence) < 6:
        suffixes = ('MR', 'MRS', 'MISS', 'MS', 'DR')  # Titles to check for
        for suffix in suffixes:
            if input_string.endswith(suffix):
                title = suffix
                break
        if title is not None:
            postcode = input_string.rstrip(title)
        else:
            postcode = input_string
    else:    
        dob = longest_sequence[-6:]
        postcode = input_string[:start_index]
        title = input_string[end_index:]
        
    
    # print('Postcode:', before, 'DOB:', last_six_numbers, 'TITLE:', after)
    
    return (postcode, dob, title)

# Create a new DataFrame without the 'String' column
postcode_dob_title_df = df.drop(columns=['String'])

# Apply the custom function to the 'String' column and create new columns in the new DataFrame
postcode_dob_title_df[['Postcode', 'DOB', 'Title']] = df['String'].apply(extract_longest_consecutive_numbers).apply(pd.Series)

```

```python
import re
import pandas as pd

# Initialize lists to store the extracted data
first_name = []
second_name = []

# Read the data from the 'officer_snapshot.txt' file
with open('officer_snapshot.txt', 'r') as file:
    for line in file:
        # Define a regular expression pattern to match text between angle brackets
        pattern = r'<([^<]+)<([^<]+)<<'

        # Find the matches in the current line
        matches = re.search(pattern, line)

        if matches:
            # Extract the two parts and append to the respective lists
            first_name.append(matches.group(1))
            second_name.append(matches.group(2))
        else:
            print (line)

# Create a DataFrame from the extracted data
names_df = pd.DataFrame({'First Name': first_name, 'Second Name': second_name})

# Display the DataFrame
# print(names_df)
```

```python
import re
import pandas as pd

# Initialize an empty list to store the parsed addresses
addresses = []

# Read the data from the 'officer_snapshot.txt' file
with open('officer_snapshot.txt', 'r') as file:
    lines = file.readlines()

# Define a regular expression pattern to extract text after '<<'
pattern = r'<<(.*)'

# Loop through the lines and extract the desired information using the pattern
for line in lines:
    match = re.search(pattern, line)
    if match:
        data = match.groups()
        # Create a dictionary with field names for this pattern
        record = {
            "Address": data[0],
        }
        addresses.append(record)  # Append the record to the list
    else:
        print(line)

# Check if any addresses were parsed
if addresses:
    addresses_df = pd.DataFrame(addresses)
    # Display the resulting DataFrame
    # print(addresses_df)
else:
    print("No addresses were parsed.")
```
